```python
project1 = pd.read_csv('/content/data_offers.csv')
project1
```

```python
project1.head()
```

```python
df = project.merge(right=project1,how='inner',on='order_gk')
df
```

```python
import numpy as np
yes_no = np.where(df['is_driver_assigned_key']==1,'yes','no')
```

```python
df['is_driver_assigned'] = yes_no
df['is_driver_assigned']
```

```python
status = np.where(df['order_status_key']==4,'client_cancel','System_cancel')
```

```python
df['status_right'] = status
df['status_right']
```

```python
df.drop(columns=['is_driver_assigned_key','order_status_key'],inplace=True)
```

```python
df = df.rename(columns={"order_datetime":"order_TIme"})
```

```python
df.sample(n=10, random_state=42)
```

```python
df_q1=df.pivot_table(columns=['is_driver_assigned','status_right'],values ='order_gk',aggfunc='count')
```

```python
df['order_hours'] = project['order_datetime'].apply(lambda x: str(x).split(':')[0])
df[['order_gk','order_hours']].head()
```

```python
df_hours = df.groupby(['order_hours'])['order_gk'].count()
df_hours
```

```python
df_hours.plot()
```

```python
df_hours.plot(xticks=range(0,24))
```

```python
df_hr = df.groupby(['order_hours','status_right'])['order_gk'].count()
df_hr.head()
```

```python
df_hr_nor = df_hr.reset_index()
df_hr_nor.head()
```

```python
df_q2_pivot = df_hr_nor.pivot(index='order_hours',columns='status_right',values='order_gk')
df_q2_pivot.head()
```

```python
df_q2_pivot.plot(xticks=range(0,24))
```

```python
df_q3 = df.groupby(by=['order_hours', 'is_driver_assigned'])['cancellations_time_in_seconds'].count()
df_q3 = df_q3.reset_index()
df_q3 = df_q3.pivot(index='order_hours', columns='is_driver_assigned', values='cancellations_time_in_seconds')
df_q3.head()
```

```python
df_q3.plot(xticks=range(0,24),figsize=(13,5))
```
